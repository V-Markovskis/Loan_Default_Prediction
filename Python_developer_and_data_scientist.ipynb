```python
import pandas as pd
df = pd.read_excel('all_data_set.xlsx', index_col = 0)
# put arguments into head for more table elements
df.head()
```

```python
# Check all columns
df.info()
```

```python
# drop redundant colums
cols = ['app_date', 'email', 'employment_status', 'employment_status', 'dob',
        'provider_1_first_seen', 'provider_1_avg_current_actual_debt', 'provider_1_last_loan_created',
        'provider_3_last_paid_debt_date', 'provider_5_device_browser_lang', 'provider_5_device_browser_type',
        'provider_5_device_firstseen', 'provider_5_device_os', 'provider_5_device_screen', 'provider_5_device_type',
        'provider_5_rule_zero', 'provider_5_rule_one', 'provider_5_rule_two', 'provider_5_rule_two',
        'provider_5_rule_three', 'provider_5_device_browser_charset', 'provider_1_avg_loan_amount_ever']
df_filtered = df.drop(cols, axis = 1)
```

```python
# check columnns after drop
df_filtered.info()
```

```python
# all columns with type object
columns_to_convert = ['provider_1_avg_rejected_apps', 'provider_1_avg_current_debt', 'provider_1_avg_max_delay', 'provider_1_avg_extensions', 'provider_1_avg_current_delay']
selected_columns_df = df_filtered[columns_to_convert]
```

```python
import numpy as np

for column in selected_columns_df:
    # Replace datetime values with NaN using regex
    df_filtered[column] = df_filtered[column].replace({r'\d{4}-\d{2}-\d{2} \d{2}:\d{2}:\d{2}': np.nan}, regex=True)

# Covert column to numberic type
for column in selected_columns_df:
    df_filtered[column] = pd.to_numeric(df_filtered[column], errors='coerce')

# Fill missing values with interpolation
for column in selected_columns_df:
    df_filtered[column] = df_filtered[column].interpolate(method='linear')
```

```python
# check if first element of column is NaN
print(df_filtered['provider_1_avg_rejected_apps'].tolist())
```

```python
# if NaN is first element, set it with mean() method from pandas
for column in selected_columns_df:
    # Ð—Ð°Ð¼ÐµÐ½Ð¸Ñ‚ÑŒ Ð¿Ñ€Ð¾Ð¿ÑƒÑ‰ÐµÐ½Ð½Ñ‹Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ñ ÑÑ€ÐµÐ´Ð½Ð¸Ð¼Ð¸ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸ÑÐ¼Ð¸
    mean_value = df_filtered[column].mean()
    df_filtered[column] = pd.to_numeric(df_filtered[column], errors='coerce').fillna(mean_value)
```

```python
# check if first element of column is still NaN
print(df_filtered['provider_1_avg_rejected_apps'].tolist())
```

```python
# check column values
df_filtered.info()
```

```python
# check boolean columns
boolean_columns = ['accept_marketing', 'ignore_communication', 'identified_by_provider_6']
selected_boolean_columns = df_filtered[boolean_columns]
for column in boolean_columns:
    print(f"Unique values in {column}: {df_filtered[column].unique()}")
```

```python
# set boolean types to numberic
for column in selected_boolean_columns:
  df_filtered[column] = df_filtered[column].astype('int64')
```

```python
# check if boolean is num now
print(df_filtered['accept_marketing'].tolist())
```

```python
df_filtered.info()
```

```python
# interpolate missing values
for column in df_filtered.columns:
  df_filtered[column] = df_filtered[column].interpolate(method='linear')
```

```python
# check if still missing values
df_filtered.info()
```

```python
# drop column which should be used for machine learning
X = df_filtered.drop('fpd30', axis = 1)
y = df_filtered['fpd30']
```

```python
# use library for machine learning
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
```

```python
# Model training
model.fit(X, y)
```
